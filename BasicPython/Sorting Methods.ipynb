```python
import random

class SortingMethods:
    
    def bubbleSort(self,_list):#冒泡排序,比较临近两个elem,交换位置O(n**2)
        for i in range(len(_list)):
            for j in range(i,len(_list)):
                if _list[i]>_list[j]:
                    _list[i],_list[j]=_list[j],_list[i]
        return _list
    
    def selectionSort(self, _list):#选择排序,每次选择最小值,最后交换前面元素和最小值的位置O(n**2)
        for i in range(len(_list)):
            min_index=i
            for j in range(i,len(_list)):
                if _list[i]>_list[j]:
                    min_index=j
            _list[i],_list[j]=_list[j],_list[i]
        return _list
    
    def inSort(self,_list):# 插入排序, 跟抓扑克一样,设置第一个元素为已排序元素,比这个元素小的元素放左边,大的放右边. O(n**2)
        finishList=[_list[0]]
        for i in range(1,len(_list)):
            breakFlag=0
            for j in range(len(finishList)):
                if _list[i]<finishList[j]:
                    finishList.insert(j,_list[i])
                    breakFlag=1
                    break
            if not breakFlag: finishList.append(_list[i])
        return finishList
    
    def poilSort(self, _list):# 桶排序 以最小值和最大值为边界创建桶,然后对每个元素计数,最后建立一个新的数组O(M+N)
        book=[0]*1001#因为这是randint(0,1000)
        for i in _list:
            book[i]+=1
        res=[]
        for i in range(len(book)):
            while book[i]:
                res.append(i)
                book[i]-=1
        return res
    
    def quickSort(self, _list):#快排 http://developer.51cto.com/art/201403/430986.htm
        self.ql=_list
        left=0
        right=len(_list)-1
        self._qs(left,right)
        return _list
    
    def _qs(self,left,right):
        standard=left
        left_pointer=left
        right_pointer=right
        if left>right:
            return
        while left!=right:
            while self.ql[right]>=self.ql[standard] and right>left:
                right-=1
            while self.ql[left]<self.ql[standard] and right>left:
                left+=1
            self.ql[left],self.ql[right]=self.ql[right],self.ql[left]
#         if left==right:
        self.ql[standard],self.ql[left]=self.ql[left],self.ql[standard]
        self._qs(left_pointer,left-1)
        self._qs(right+1,right_pointer)
            
    def mergeSort(self, _list):# 并归 把原来的列表打散,然后依次重组. 类似于divide and conquer
        if len(_list)<=1:
            return _list
        mid=len(_list)//2
        left=self.mergeSort(_list[:mid])
        right=self.mergeSort(_list[mid:])
        return self._merge(left,right)
        
    def _merge(self, left,right):
        res=[]
        leftPointer=rightPointer=0
        while leftPointer<len(left) and rightPointer<len(right):
            if left[leftPointer]<right[rightPointer]:
                res.append(left[leftPointer])
                leftPointer+=1
            else:
                res.append(right[rightPointer])
                rightPointer+=1                
        if leftPointer<len(left):
            res+=left[leftPointer:]
        if rightPointer<len(right):
            res+=right[rightPointer:]
        return res
    
    def shellSort(self, _list):# 
        """
        图解的话看 https://www.itcodemonkey.com/article/3286.html
        但代码是傻逼版本,理论上不需要创建这么多中间list,导致空间复杂度增大.但这确实相对比较好理解
        https://www.jianshu.com/p/b92d791ad008
        
        希尔排序用到了插入排序.
        时间复杂度很复杂,
        大概是O(n**1.3)
        
        https://www.itcodemonkey.com/article/3291.html
        时间复杂度的傻瓜版解释
        """
        gap=len(_list)
        while gap>=0:
            temp_list=[]
            if gap==0:
                return _list
            for index, value in enumerate(_list):
                if index%gap ==0:
                    catch=[index,value]
                    temp_list.append(catch)
            temp_value_list=[]
            for i in temp_list:
                temp_value_list.append(i[1])
            temp_value_list=self.inSort(temp_value_list)
            for i, vv in enumerate(temp_value_list):
                temp_list[i][1]=vv
            for iv in temp_list:
                _list[iv[0]]=iv[1]
            gap=gap//2
        return _list
            
            
        
            
if __name__=='__main__':
    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().bubbleSort(a)==sorted(a)
    
    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().bubbleSort(a)==sorted(a)

    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().inSort(a)==sorted(a)

    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().poilSort(a)==sorted(a)
    
    a = [random.randint(0,1000) for _ in range(10)]
    assert SortingMethods().quickSort(a[:])==sorted(a[:])
    
    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().mergeSort(a)==sorted(a)
    
    a = [random.randint(0,1000) for _ in range(100)]
    assert SortingMethods().shellSort(a)==sorted(a)
    print("ALL CORRECT!")
```

```python
#quickSort
def quickSort(input):
    nums=input[:]
    qs(nums,0, len(nums)-1)
    return nums

def qs(nums, leftPoint, rightPoint):
    if leftPoint>=rightPoint:
        return
    l=leftPoint
    r=rightPoint
    std=l
    while l!=r:
        while r>l and nums[r]>nums[std]:
            r-=1
        while r>l and nums[l]<=nums[std]:
            l+=1
        nums[l], nums[r]= nums[r], nums[l]
    nums[l], nums[std]=nums[std],nums[l]
    qs(nums,leftPoint,l-1)
    qs(nums,r+1,rightPoint)
    

import random

a=[random.randint(0,1000) for _ in range(100)]
test=quickSort(a)
print(test==sorted(a))
```

```python
#bubbleSort

def Bubble(nums):
    for i in range(len(nums)-1):
        for j in range(i+1,len(nums)):
            if nums[i]>nums[j]:
                nums[i],nums[j]=nums[j],nums[i]
    return nums
import random

a=[random.randint(0,1000) for _ in range(100)]
test=Bubble(a)
print(test==sorted(a))              
```
