## 515. Find Largest Value in Each Tree Row


[515. Find Largest Value in Each Tree Row](https://leetcode.com/problems/find-largest-value-in-each-tree-row/#/description)


BFS

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root is None:
            return []
        stack=[root]
        res=[]
        while stack:
            temp=stack
            stack=[]
            level_max=float('-inf')
            for node in temp:
                if node:
                    if node.val>level_max:
                        level_max=node.val
                    if node.left:
                        stack.append(node.left)
                    if node.right:
                        stack.append(node.right)
            res.append(level_max)
        return res
        
```

## 236. Lowest Common Ancestor of a Binary Tree

[236. Lowest Common Ancestor of a Binary Tree](https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-tree/#/description)


分治的方法, 
* 好好看看~

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
#         divide
        if not root:
            return None
        if root == p or root == q:
            return root
#         conquer
        right=self.lowestCommonAncestor(root.right, p, q)
        left=self.lowestCommonAncestor(root.left, p, q)
        
        if right and left:
            return root
        if right: return right
        if left: return left

        
```

## 199. Binary Tree Right Side View

[199. Binary Tree Right Side View](https://leetcode.com/problems/binary-tree-right-side-view/#/description)

递归的方法就是查层数,然后用append每层第一个数.

循环的话,就是用队列做,BFS,记录每排第一个数字.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def rightSideView(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root:
            return []
        self.res=[]
        
        def dfs(node, level):
            if not node:
                return 
            if level == len(self.res):
                self.res.append(node.val)
            dfs(node.right,level +1)
            dfs(node.left, level+1)
        dfs(root,0)
        return self.res

    def rightSideView1(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root is None:
            return []
        queue=[root]
        res=[]
        
        while queue:
            sub_queue=queue[:]
            level_list=[]
            queue=[]
            while sub_queue:
                node=sub_queue.pop(0)
                level_list.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            res.append(level_list.pop())
        return res        
```

## 114. Flatten Binary Tree to Linked List

[114. Flatten Binary Tree to Linked List](https://leetcode.com/problems/flatten-binary-tree-to-linked-list/#/solutions)

递归DFS. 

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def __init__(self):
        self.prev=None
    
    def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: void Do not return anything, modify root in-place instead.
        """
        if root is None:
            return
        self.flatten(root.right)
        self.flatten(root.left)
        
        root.right=self.prev# 每个节点的右子叶设置为之前的节点prev
        root.left=None# 左子叶设置为None
        self.prev=root# 然后把该节点设置为prev
        
```

## 230. Kth Smallest Element in a BST

[230. Kth Smallest Element in a BST](https://leetcode.com/problems/kth-smallest-element-in-a-bst/#/description)

就是遍历二叉树. 

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def kthSmallest(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: int
        """
        self.k=k
        self.res=0
        def dfs(node):
            if node is None:
                return
            dfs(node.left)
            self.k-=1
            if self.k == 0:
                self.res=node.val
                return 
            dfs(node.right)
        dfs(root)
        return self.res
                
```
