## 21. Merge Two Sorted Lists



[21. Merge Two Sorted Lists](https://leetcode.com/problems/merge-two-sorted-lists/description/)
指针基础



```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        head=ListNode(None)
        res=head
        while l1 and l2:
            if l1.val<l2.val:
                res.next=l1
                l1=l1.next
            else:
                res.next=l2
                l2=l2.next
            res=res.next#  每次循环结束向前移动指针
        if l1:
            res.next=l1
        elif l2:
            res.next=l2
        return head.next
```

## 86. Partition List



[86. Partition List](https://leetcode.com/problems/partition-list/description/)
双LL,最后首位组合,输出小LL的头

build two link list and link small tail and not less head

注意:
*  用curr.next赋值(line 17,19)
*  结束时设置otherLL末尾为None,否则容易出现循环数组(line 25)


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def partition(self, head, x):
        """
        :type head: ListNode
        :type x: int
        :rtype: ListNode
        """
        small_head,other_head=ListNode(None),ListNode(None)
        small_curr,other_curr=small_head,other_head
        curr=head
        while curr!=None:
            if curr.val<x:
                small_curr.next=curr
                small_curr=small_curr.next
            else:
                other_curr.next=curr
                other_curr=other_curr.next
            curr=curr.next
        other_curr.next=None
        small_curr.next=other_head.next
        return small_head.next
```

## 141. Linked List Cycle



[141. Linked List Cycle](https://leetcode.com/problems/linked-list-cycle/description/)

快慢指针
检查有没有闭环,就是看快慢指针是否追尾

注意:
* 检查fast and fast.next 而不是fast.next.next(不知道为什么!)(line 18)

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        if head==None or head.next==None or head.next.next==None:
            return False
        slow=head.next
        fast=head.next.next
        
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next
            if fast == slow:
                return True
        return False
```

## 160. Intersection of Two Linked Lists



[160. Intersection of Two Linked Lists](https://leetcode.com/problems/intersection-of-two-linked-lists/description/)

本题有几种解法:
* 查链表长度,然后补齐短的链表(长链表先走几步). 双指针一起前进,遇到相同的就True
* 两链表首尾相接,然后用快慢指针做,最后恢复链表

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        lA,lB=0,0
        curA,curB=headA,headB
        while curA!=None:
            curA=curA.next
            lA+=1
        while curB!=None:
            curB=curB.next
            lB+=1
        curA,curB=headA,headB
        s=lA-lB
        if s>=0:
            while s:
                curA=curA.next
                s-=1
        else:
            while s:
                curB=curB.next
                s+=1
        while curA!=None:
            if curA==curB:
                return curA
            curA=curA.next
            curB=curB.next
```

## 203. Remove Linked List Elements



[203. Remove Linked List Elements](https://leetcode.com/problems/remove-linked-list-elements/description/)

两种玩法:

* 新建一个LL, 然后遍历之前的LL,凡是node.val==target的,就不加入到新的LL中,最后封口新的LL然后输出

* 建立prev和dummy,dummy用于保存new_head,防止第一个元素被删除,每次检查时前进prev和curr,当curr.val==val时,prev前进两格(line 40!)


```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def removeElements1(self, head, val):# 这是第一种解法
        """
        :type head: ListNode
        :type val: int
        :rtype: ListNode
        """
        res=ListNode(None)
        curr=res
        while head:
            if head.val==val:
                head=head.next
                continue
            else:
                curr.next=head
                curr=curr.next
            head=head.next
        curr.next=None
        return res.next
    
    def removeElements2(self, head, val):#这是第二种解法
        """
        :type head: ListNode
        :type val: int
        :rtype: ListNode
        """
        prev=ListNode(-1)
        prev.next=head
        curr=head
        dummy=prev
        
        while curr:
            if curr.val==val:
                prev.next=curr.next
            else:
                prev=prev.next
            curr=curr.next
        return dummy.next
```

## 206. Reverse Linked List



[206. Reverse Linked List](https://leetcode.com/problems/reverse-linked-list/description/)


翻转链表

迭代和递归,两种玩法
递归有点不太明白~

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def reverseList1(self, head):#迭代,也就是循环
        """
        :type head: ListNode
        :rtype: ListNode
        """
        prev=None
        curr=head
        while curr:
            next1=curr.next#先留下一个的种子
            curr.next=prev#把指针指向前
            prev=curr#prev接收新的curr值
            curr=next1#curr向后移动
        return prev
    
    def reverseList2(self, head):#递归
        return self._reverse(head)

    def _reverse(self, node, prev=None):
        if not node:
            return prev#结尾后返回
        n = node.next#这是原LL的下一个节点
        node.next = prev#指向之前的点
        return self._reverse(n, node)
```

## 234. Palindrome Linked List



[234. Palindrome Linked List](https://leetcode.com/problems/palindrome-linked-list/description/)


* 快慢指针, 翻转后半部分链表
* 或者用栈来做,前半部分压入栈,然后pop对比和后半部分的区别

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def isPalindrome(self, head):# 空间复杂度是O(1) 翻转链表
        """
        :type head: ListNode
        :rtype: bool
        """
        slow=head
        fast=head
        length=0
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next
        #reverse
        prev=None
        curr=slow
        while curr:
            next1=curr.next
            curr.next=prev
            prev=curr
            curr=next1
        #cmp
        while prev:
            if prev.val!=head.val:
                return False
            prev=prev.next
            head=head.next
        return True
```

## 83. Remove Duplicates from Sorted List



[83. Remove Duplicates from Sorted List](https://leetcode.com/problems/remove-duplicates-from-sorted-list/description/)

有序链表去重

注意:
* 用prev.next来设置指向,而不是prev


```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        dummy=ListNode(None)
        dummy.next=head
        curr=head
        prev=dummy        
        while curr:
            if curr.val==prev.val:
                prev.next=curr.next
            else:
                prev=prev.next
            curr=curr.next
        return dummy.next
```
