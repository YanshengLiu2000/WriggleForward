## 110. Balanced Binary Tree

平衡二叉树

[110. Balanced Binary Tree](https://leetcode.com/problems/balanced-binary-tree/description/)

用递归和深度优先做. 这就是熟悉递归和深度优先的题


```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isBalanced(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        
        self.result=True
        self.dfs(root)
        return self.result

    
    def dfs(self, node):
        if node is None:
            return 0
        left=self.dfs(node.left)
        right=self.dfs(node.right)
        if abs(left-right)>1:
            self.result=False
        return max(left,right)+1
    
```

## 107. Binary Tree Level Order Traversal II

二叉树 遍历

[107. Binary Tree Level Order Traversal II](https://leetcode.com/problems/binary-tree-level-order-traversal-ii/description/)


本题有两种玩法~可以用栈BFS 或者用递归BFS

* 栈: 在初始化时压入root,之后每层一次导出node, 压入对应的下一层. 最后用切片来颠倒整个list

* 递归: 在每次调用函数的时候检查该层是否已经建立子数组,然后append入该点val. 最后递归入左右子树


```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def levelOrderBottom(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if root is None:
            return []
        self.res=[]     
        stack=[root]
        while stack:
            level_nodes=stack
            stack=[]
            temp=[]
            for node in level_nodes:
                temp.append(node.val)
                if node.left:
                    stack.append(node.left)
                if node.right:
                    stack.append(node.right)
            self.res.append(temp)
        return self.res[::-1]
    
    def levelOrderBottom2(self, root):# 递归的方法
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        self.res=[]
        self.helper(root,0)
        return self.res[::-1]
    
    def helper(self, node, level):
        if node is None:
            return
        if len(self.res)== level:
            self.res.append([])
        self.res[level].append(node.val)
        self.helper(node.left,level+1)
        self.helper(node.right, level+1)
```

## 112. Path Sum

二叉树 遍历

[112. Path Sum](https://leetcode.com/problems/path-sum/description/)


* 递归的思路, 结束信号就是检查根节点,一旦检测到根节点就判断total==sum. 

* 栈的思路, 先进后出,栈内压入(node, total), 判断根节点.

栈的方法比递归要慢很多多, 不知道为什么.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def hasPathSum(self, root, sum):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: bool
        """
        if root is None:
            return False
        self.res=False
        self.helper(root, root.val, sum)
        return self.res
        
    def helper(self, node, total, sum):
        if not node.left and not node.right and total == sum:
            self.res=True
            return
        if node.left:
            self.helper(node.left,total+node.left.val, sum)
        if node.right:
            self.helper(node.right, total+node.right.val, sum)
    
    def hasPathSum2(self, root, sum):#栈的方法
        """
        :type root: TreeNode
        :type sum: int
        :rtype: bool
        """
        if root is None:
            return False
        # res=False
        stack=[(root,0)]
        while stack:
            node,total=stack.pop()
            if not node.left and not node.right and total+node.val==sum:
                return True
            if node.left:
                stack.append((node.left, total+node.val))
            if node.right:
                stack.append((node.right, total+node.val))
        return False
```

## 104. Maximum Depth of Binary Tree

二叉树 遍历

[104. Maximum Depth of Binary Tree](https://leetcode.com/problems/maximum-depth-of-binary-tree/)

用的宽度优先搜索,和递归.
也可以用深度优先做.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        def func(depth, node):
            if node is None:
                return 1
            depth=depth+max(func(depth,node.left),func(depth,node.right))+1
            return depth
        res=func(0, root)-1
        return res
    def maxDepth1(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return 0
        depth=0
        stack=[root]
        while stack:
            depth+=1
            temp_stack=stack[:]
            stack=[]
            while temp_stack:
                
                node=temp_stack.pop()
                if node.left:
                    stack.append(node.left)
                if node.right:
                    stack.append(node.right)
        return depth
```

## 108. Convert Sorted Array to Binary Search Tree

构建平衡二叉树

[108. Convert Sorted Array to Binary Search Tree](https://leetcode.com/problems/convert-sorted-array-to-binary-search-tree/description/)

递归的方法, 再加上用python的切片完成.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def sortedArrayToBST(self, nums):
        """
        :type nums: List[int]
        :rtype: TreeNode
        """
        if nums is None or nums == []:
            return None
        if not len(nums)%2:
            mid=len(nums)//2-1
        else:
            mid=len(nums)//2
        root=TreeNode(nums[mid])
        root.left=self.sortedArrayToBST(nums[:mid])
        root.right=self.sortedArrayToBST(nums[mid+1:])
        return root
```

## 111. Minimum Depth of Binary Tree

构建平衡二叉树

[111. Minimum Depth of Binary Tree](https://leetcode.com/problems/minimum-depth-of-binary-tree/#/description)

用栈和BFS的方法相对较慢

递归的话相对快很多, 正常计算depth用max(node.left, node.right), 这里由于是要最小深度,就用min(node.left, node.right)

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return 0
        stack=[root]
        depth=0
        while stack:
            temp=stack
            depth+=1
            stack=[]
            while temp:
                node=temp.pop()
                if node.left:
                    stack.append(node.left)
                if node.right:
                    stack.append(node.right)
                if not node.left and not node.right:
                    return depth
    def minDepth1(self, root):
         if not root: 
            return 0;
        left = self.minDepth(root.left)
        right = self.minDepth(root.right)
        if left == 0 or right == 0:
            return left + right + 1
        return min(left, right) + 1           
        
```

## 235. Lowest Common Ancestor of a Binary Search Tree



[235. Lowest Common Ancestor of a Binary Search Tree](https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-search-tree/#/description)

迭代,

当node.val比两个数都小时说明node太小,应该增大node, node=node.right

当node.val比两个数都大时说明node太大,应该减小node, node= node.left

剩下的情况就是node与期中一个相等或者正好在两个数中间,那就返回这个node.

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if root is None or not q or not p:
            return None
        node=root
        while node:
            if node.val>max(p.val,q.val):
                node=node.left
            elif node.val<min(p.val,q.val):
                node=node.right
            else:
                return node
    def lowestCommonAncestor1(self, root, p, q):# 递归的方法
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        if root is None or not q or not p:
            return None
        if root.val>max(p.val, q.val):
            return self.lowestCommonAncestor(root.left, p, q)# 小心这一步之, 直接返回 func就可以了, 不需要设定root=什么
        if root.val<min(p.val, q.val):
            return self.lowestCommonAncestor(root.right, p, q)
        else:
            return root
```

## 226. Invert Binary Tree



[226. Invert Binary Tree](https://leetcode.com/problems/invert-binary-tree/#/description)

BFS, 广度优先搜索, 对所有的点都有: 左右子树交换

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def invertTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        if root is None:
            return 
        queue=[root]
        while queue:
            temp=queue
            queue=[]
            while temp:
                node=temp.pop()
                if node:
                    node.left, node.right=node.right, node.left
                    queue.append(node.left)
                    queue.append(node.right)
        return root
```

## 543. Diameter of Binary Tree



[543. Diameter of Binary Tree](https://leetcode.com/problems/diameter-of-binary-tree/#/description)

递归, 

* 再看看

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def diameterOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return 0
        self.max=0
        
        def curmax(node):
            if node is None:
                return 0
            left=curmax(node.left)
            right=curmax(node.right)
            
            self.max=max(self.max, left + right)
            
            return 1+max(left,right)
        curmax(root)
        return self.max
```

## 501. Find Mode in Binary Search Tree



[501. Find Mode in Binary Search Tree](https://leetcode.com/problems/find-mode-in-binary-search-tree/#/description)

找众数

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def findMode(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if root is None:
            return []
        def buildHash(hash, node):
            if node is None:
                return
            if node.val in hash:
                hash[node.val]+=1
            else:
                hash[node.val]=1
            buildHash(hash, node.left)
            buildHash(hash, node.right)
        
        hash={}
        buildHash(hash,root)
        max_val=max(hash.values())
        res=[]
        for key in hash:
            if hash[key]==max_val:
                res.append(key)
        return res
```

## 257. Binary Tree Paths



[257. Binary Tree Paths](https://leetcode.com/problems/binary-tree-paths/#/description)

就是DFS

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def binaryTreePaths(self, root):
        """
        :type root: TreeNode
        :rtype: List[str]
        """
        if root is None:
            return []
        stack=[(str(root.val),root)]
        res=[]
        while stack:
            string, node=stack.pop()
            if node.left:
                temp=string+'->'+str(node.left.val)
                stack.append((temp,node.left))
            if node.right:
                temp=string+'->'+str(node.right.val)
                stack.append((temp,node.right))
            if node.left is None and node.right is None:
                res.append(string)
        return res
                
```

## 572. Subtree of Another Tree



[572. Subtree of Another Tree](https://leetcode.com/problems/subtree-of-another-tree/#/description)

对s树的每个点进行比较, 但不知道怎么回事,无法通过全部测试.

```python
class Solution:
    def isSubtree(self, s, t):
        """
        :type s: TreeNode
        :type t: TreeNode
        :rtype: bool
        """
        def isMatch(s,t):
            if s is None and t is None:
                return True
            if s is None or t is None:
                return False
            if s.val == t.val:
                return isMatch(s.left, t.left) and isMatch(s.right, t.right)
            else:
                return False
        if isMatch(s,t):
            return True
        else:
            return isMatch(s.left,t) or isMatch(s.right,t)
        
```

## 404. Sum of Left Leaves



[404. Sum of Left Leaves](https://leetcode.com/problems/sum-of-left-leaves/#/description)


找到左子叶.

本题中递归的方法更快~

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def sumOfLeftLeaves(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None:
            return 0
        stack=[(0,root)]
        res=0
        while stack:
            tag,node=stack.pop()
            if tag and node.left is None and node.right is None:
                res+=node.val
            if node.left:
                stack.append((1,node.left))
            if node.right:
                stack.append((0, node.right))
        return res
    
    def sumOfLeftLeaves1(self, root):# 递归, 本题递归的方法更快
        """
        :type root: TreeNode
        :rtype: int
        """
        # Edge
        if not root:
            return 0
        # Process
        self.count = 0
        def dfs(root):
            if not root:
                return 0
            if root.left and not root.left.left and not root.left.right:
                self.count += root.left.val
            # Recursion
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return self.count
        
```

## 563. Binary Tree Tilt



[563. Binary Tree Tilt](https://leetcode.com/problems/binary-tree-tilt)


不是很明白为什么需要返回node.val+left+right

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def findTilt(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.res=0
        
        def helper(node):
            if node is None:
                return 0
            left=helper(node.left)
            right=helper(node.right)
            self.res+=abs(left-right)
            return node.val+left+right
        helper(root)
        return self.res
```

## 637. Average of Levels in Binary Tree



[637. Average of Levels in Binary Tree](https://leetcode.com/problems/average-of-levels-in-binary-tree/)


这个题不难,就是基本的BFS

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def averageOfLevels(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        if root is None:
            return 0
        stack=[root]
        res=[]
        while stack:
            length=len(stack)
            temp=stack
            stack=[]
            total=0
            while temp:
                node=temp.pop()
                if node.left:
                    stack.append(node.left)
                if node.right:
                    stack.append(node.right)
                total+=node.val
            res.append(total*1.0/length)
        return res
                    
        
```

## 538. Convert BST to Greater Tree



[538. Convert BST to Greater Tree](https://leetcode.com/problems/convert-bst-to-greater-tree/#/description)


这个也不难,就是基本的中序遍历.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def convertBST(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        self.sum=0
        def dfs(node):
            if node is None:
                return 0
            dfs(node.right)
            node.val+=self.sum
            self.sum=node.val
            dfs(node.left)
        dfs(root)
        return root
            
        
```
