# BackTracking

## 78. Subsets

全组和

[78.Subsets](https://leetcode.com/problems/subsets/description/)
本题用树(DFS)的思想解决

进入dfs 后直接append 切片就好了


```python
class Solution:
    def subsets(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        self.res=[]
        
        def dfs(nums,index,temp):
            self.res.append(temp[:])
            for i in range(index,len(nums)):
                temp.append(nums[i])
                dfs(nums,i+1,temp)
                temp.pop()
            
        dfs(nums,0,[])
        return self.res
```

## 90.Subsets II

全组和

[90.Subsets II](https://leetcode.com/problems/subsets-ii/description/)
本题和上一道 78题差不多, 在self.res.append前需要复制一份temp. 检查是否已经存在. 
如果不存在的话, append如temp[:]


```python
class Solution:
    def subsetsWithDup(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        self.res=[]
        
        def dfs(nums, index, temp):
            # # temp.sort()
            # if temp not in self.res:
            #     self.res.append(temp[:])
            check_temp=sorted(temp)
            if check_temp not in self.res:
                self.res.append(check_temp[:])
            for i in range(index,len(nums)):
                temp.append(nums[i])
                dfs(nums, i+1, temp)
                temp.pop()
            
        dfs(nums, 0, [])
        self.res=sorted(self.res)
        return self.res
```

## 46. Permutations

[46. Permutations](https://leetcode.com/problems/permutations/description/)


全排列

解题思想与全组和没有任何不同, 本质上来说还是对树结构的DFS. 每次遍历整个list,遇到重复的就跳过append. 当遇到根节点后append入self.res

_可以改进优化算法._



```python
class Solution:
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        self.res=[]
        
        def dfs(nums, temp):
            if len(temp) == len(nums):
                self.res.append(temp[:])
            for i in range(len(nums)):
                if nums[i] in temp:
                    continue
                temp.append(nums[i])
                dfs(nums,temp)
                temp.pop()
                
        dfs(nums,[])
        return self.res
    
```

## 60. Permutation Sequence

[60. Permutation Sequence](https://leetcode.com/problems/permutation-sequence/description/)

全排列

没有重复数字的全排列. 

本质上和[46. Permutations](https://leetcode.com/problems/permutations/description/)没有不同.

但运算时间会超时.

```python
class Solution:
    def getPermutation(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: str
        """
        self.res=[]
        self.nums=[i+1 for i in range(n)]
        
        def dfs(temp):
            if len(temp) == len(self.nums):
                self.res.append(temp[:])
            for i in range(len(self.nums)):
                if self.nums[i] in temp:
                    continue
                temp.append(self.nums[i])
                dfs(temp)
                temp.pop()
        
        dfs([])
        string=''
        for i in self.res[k-1]:
            string+=str(i)
        return string
```

## 47. Permutations II

[47. Permutations II](https://leetcode.com/problems/permutations-ii/description/)

全排列

有重复数字的全排列. 

先排序, 然后建立first_position list 用于检测是否首位用过. 用usedlist存储是否当前用过, 再pop 后,used list 对应元素需要转成True. 在检查是否合法append入res时,检查len(temp)和temp是否存在在res中.

本质上和全组和一样.

```python
class Solution:
    def permuteUnique(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        self.res=[]
        self.check=[False]*len(nums)
        self.used=[False]*len(nums)
        self.first_position=[]
        nums=sorted(nums)
   
        def dfs(nums,temp):
            if len(nums) == len(temp) and temp not in self.res:
                self.res.append(temp[:])
            for i in range(len(nums)):
                if temp==[]:
                    if nums[i] in self.first_position:
                        continue
                    else:
                        self.first_position.append(nums[i])
                if self.used[i]:
                    continue
                self.used[i]=True
                temp.append(nums[i])
                dfs(nums,temp)
                self.used[i]=False
                temp.pop()
                    
        dfs(nums,[])
        return self.res
```

## 77. Combinations

[47. Permutations II](https://leetcode.com/problems/combinations/description/)

有限全排列

指定位数的全排列

这样在DFS时不需要查到根节点,当长度满足要求时,直接append, 长度大于要求时,返回上一层

```python
class Solution:
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """
        self.res=[]
        nums=[i+1 for i in range(n)]
        
        def dfs(nums, index, temp, k):
            if len(temp)==k:
                self.res.append(temp[:])
            elif len(temp)>k:
                return
            for i in range(index,len(nums)):
                temp.append(nums[i])
                dfs(nums,i+1,temp,k)
                temp.pop()
        dfs(nums,0,[],k)
        return self.res
```

## 39. Combination Sum

[39. Combination Sum](https://leetcode.com/problems/combination-sum/description/)

组合加(无重复数字,但同一数字可选择多次)

def dfs(nums, target, temp,index):

仍旧使用backtracking的方法, 每次乡下传递时,index不变,这样程序会考虑使用同一个数字多次的.target-=当前数字

判断条件改成target==0 append; target<0 back



```python
class Solution:
    def combinationSum(self, nums, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.res=[]
        
        def dfs(nums, target, temp,index):
            if target==0:
                self.res.append(temp[:])
                return
            elif target <0:
                return
            
            for i in range(index,len(nums)):
                temp.append(nums[i])
                dfs(nums,target-nums[i],temp,i)
                temp.pop()
        dfs(nums, target,[], 0)
        return self.res
```

## 40. Combination Sum II

[40. Combination Sum II](https://leetcode.com/problems/combination-sum-ii/description/)

组合加(重复数字,同一数字只能用1次)

老样子,先排序

def dfs(nums, target, temp,index):

仍旧使用backtracking的方法, 每次乡下传递时,index+1,target-=nums[i]

判断条件改成target==0 and not in res, append; target<0, back

```python
class Solution:
    def combinationSum2(self, nums, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.res=[]
        nums=sorted(nums)
        def helper(nums, temp, index, target):
            # check=sorted(temp)
            if target == 0 and temp not in self.res:
                self.res.append(temp[:])
                return
            elif target < 0:
                return
            
            for i in range(index,len(nums)):
                temp.append(nums[i])
                helper(nums,temp, i+1,target-nums[i])
                temp.pop()
                
        helper(nums,[],0,target)
        return self.res
```

## 216. Combination Sum III

[216. Combination Sum III](https://leetcode.com/problems/combination-sum-iii/description/)

组合加(无重复数字,同一数字只能用1次)

和[40. Combination Sum II](https://leetcode.com/problems/combination-sum-ii/description/)
没有任何不同, 需要在解题前建立nums数组


```python
class Solution:
    def combinationSum3(self, k, n):
        """
        :type k: int
        :type n: int
        :rtype: List[List[int]]
        """
        self.res=[]
        nums=[i+1 for i in range(9)]
        
        def helper(nums, temp, index,target):
            if target == 0 and len(temp)==k:
                self.res.append(temp[:])
                return
            elif target<0:
                return
            
            for i in range(index,9):
                temp.append(nums[i])
                helper(nums,temp,i+1,target-nums[i])
                temp.pop()
            
        
        helper(nums, [], 0, n)
        return self.res
```

## 17. Letter Combinations of a Phone Number

[17. Letter Combinations of a Phone Number](https://leetcode.com/problems/letter-combinations-of-a-phone-number/description/)

全组和

还是backtracking的思想, 建立一个book={} ,遍历第一个数字对应的字母数组, 向下传递时index+1

当len(temp)== len(string)的时候append


```python
class Solution:
    def letterCombinations(self, string):
        """
        :type digits: str
        :rtype: List[str]
        """
        self.book={'2':['a','b','c'],'3':['d','e','f'],'4':['g','h','i'],'5':['j','k','l'],'6':['m','n','o'],
     '7':['p','q','r','s'],'8':['t','u','v'],'9':['w','x','y','z']}
        self.res=[]
        if len(string)==0:
            return []
        def help(string, index, temp):
            if len(string)==len(temp):
                self.res.append("".join(i for i in temp))
                return
            
            for char in self.book[string[index]]:
                temp.append(char)
                help(string, index+1,temp)
                temp.pop()
        help(string, 0, [])
        return self.res
```

## 131. Palindrome Partitioning

[131. Palindrome Partitioning](https://leetcode.com/problems/palindrome-partitioning/description/)

回文串

写一个s==s{::-1} 来判断是否是回文串

def dfs(s,temp):

对s遍历, 从1到len(s)+1 因为需要一点点截取

向下传递时传递dfs(s[i:], temp)

判断条件是没有s,append全部的temp.

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        self.res=[]
        
        def isPal(s):
            return s==s[::-1]
        
        def dfs(s,temp):
            #return condition
            if not s :
                self.res.append(temp[:])
            #backtracking
            for i in range(1,len(s)+1):
                if isPal(s[:i]):
                    temp.append(s[:i])
                    dfs(s[i:],temp)
                    temp.pop()
                
        dfs(s,[])
        return self.res
```
