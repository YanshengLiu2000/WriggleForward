## 2. Add Two Numbers



[2. Add Two Numbers](https://leetcode.com/problems/add-two-numbers/description/)

双指针进位问题

注意:
* 还是老问题,注意用curr=head; curr.next=ListNode()向下传递LL
* 检测l1, l2 和 flag,查看是否需要新建Node

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        flag=0
        head=ListNode(None)
        curr=head
        while l1 or l2 or flag:
            temp=0
            if l1:
                temp+=l1.val
                l1=l1.next
            if l2:
                temp+=l2.val
                l2=l2.next
            if flag:
                temp+=1
                flag=0
            if temp>=10:
                flag=1
                temp-=10
            curr.next=ListNode(temp)
            curr=curr.next
        return head.next
```

## 445. Add Two Numbers II



[445. Add Two Numbers II](https://leetcode.com/problems/add-two-numbers-ii/description/)

本题与上一题不同,本题用栈来解比较方便,顺序压入栈中然后依次出栈

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        s1,s2=[],[]      #新建两个对应的栈
        while l1:        #l1压栈
            s1.append(l1.val)
            l1=l1.next
        while l2:       #l2压栈
            s2.append(l2.val)
            l2=l2.next
            
        flag=0#进位符
        prev=ListNode(None)
        curr=ListNode(None)
        prev.next=curr#倒着构建一个LL
        while s1 or s2 or flag:
            temp=0
            if s1:
                temp+=s1.pop()
            if s2:
                temp+=s2.pop()
            if flag:
                temp+=1
                flag=0
            if temp>=10:
                temp-=10
                flag=1
            #目前为止与第二题一样.
            #之后可以倒着建立一个LL或者用第三个stack来存储temp结果,然后正向建立一个个LL
            curr.val=temp
            curr=prev
            prev=ListNode(None)
            prev.next=curr
        return prev.next.next
```

## 142. Linked List Cycle II

快慢指针

[142. Linked List Cycle II](https://leetcode.com/problems/linked-list-cycle-ii/description/)

本题与141大致相同[141. Linked List Cycle](https://leetcode.com/problems/linked-list-cycle/description/)

https://www.cnblogs.com/hiddenfox/p/3408931.html

查看图解

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
#         https://www.cnblogs.com/hiddenfox/p/3408931.html
        if head is None or head.next is None:
            return None 
        slow,fast = head, head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            if fast is slow:
                slow=head
                while True:
                    if fast is slow:
                        return fast
                    slow=slow.next
                    fast=fast.next

        return None
```

## 328. Odd Even Linked List

连续双指针

[328. Odd Even Linked List](https://leetcode.com/problems/odd-even-linked-list/description/)

连续双指针,一个指向基数,一个指向偶数

多理解,用数字标出来


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def oddEvenList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head:
            return head

        odd=head#1
        even=head.next#2
        evenhead=even
        
        while even and even.next:
            odd.next=even.next
            odd=odd.next
            even.next=odd.next
            even=even.next
        odd.next=evenhead
        return head
```

## 143. Reorder List

双指针

[143. Reorder List](https://leetcode.com/problems/reorder-list/description/)

仍旧是双指针问题,用慢指针截取LL的中间元素,然后反转LL的后半部分. 最后连接两个LL.酱紫~


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def reorderList(self, head):
        """
        :type head: ListNode
        :rtype: void Do not return anything, modify head in-place instead.
        """
        fast=slow=head
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next
        
        head1,head2=head,slow
        
        curr, prev=slow, None
        while curr:
            _next=curr.next
            curr.next=prev
            prev=curr
            curr=_next
        
        curr1,curr2=head1, prev
        while curr1 and curr1.next and curr2:
            _next1,_next2=curr1.next, curr2.next
            curr1.next=curr2
            curr2.next=_next1
            curr1=_next1
            curr2=_next2
```

## 61. Rotate List

环链表转向问题

[61. Rotate List](https://leetcode.com/problems/rotate-list/description/)

首尾连接后,发现规律然后转向. 转到目标点后,断开链表环.


```python
class Solution:
    def rotateRight(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if head is None or head.next==None:
            return head
        #count length
        count=1
        curr=head
        
        while curr.next:
            count+=1
            curr=curr.next
        curr.next=head
        # check rotate
        move=count-k%count+1
        #move
        prev=None
        while move:
            prev=curr
            curr=curr.next
            move-=1
        prev.next=None
        return curr
```

## 24. Swap Nodes in Pairs

指针

[24. Swap Nodes in Pairs](https://leetcode.com/problems/swap-nodes-in-pairs/description/)

首尾连接后,发现规律然后转向. 转到目标点后,断开链表环.


```python
class Solution:
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head is None or head.next is None:
            return head
        #用于保存head,返回的时候用
        dummy=ListNode(None)
        dummy.next=head
        # 建立两个变量,一个用于保存之前的点,一个用于遍历链表
        curr=head
        prev=dummy
        while curr and curr.next:
            #  prev->node1->node2->next_pair
            node1=curr
            node2=curr.next
            next_pair=node2.next
            prev.next=node2
            node2.next=node1
            node1.next=next_pair
            curr=next_pair
            prev=node1
            
        return dummy.next
```

## 148. Sort List

指针

[148. Sort List](https://leetcode.com/problems/sort-list/description/)

做法不对,但可以这么做


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        curr=head
        val_list=[]
        while curr:
            val_list.append(curr.val)
            curr=curr.next
        curr=head
        for i in sorted(val_list):
            curr.val=i
            curr=curr.next
        return head
        
```

## 19. Remove Nth Node From End of List

类似快慢指针

[19. Remove Nth Node From End of List](https://leetcode.com/problems/remove-nth-node-from-end-of-list/description/)

类似快慢指针,让快指针先走n个距离,然后双指针等速前进.

当快指针到队尾的时候,删除slow.next这个点.

```python
class Solution:
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        dummy=ListNode(None)
        dummy.next=head
        
        s=f=dummy
        for i in range(n):
            f=f.next
        while f.next:
            f=f.next
            s=s.next
        s.next=s.next.next
        return dummy.next
```

## 138. Copy List with Random Pointer

字典,链表

[138. Copy List with Random Pointer](https://leetcode.com/problems/copy-list-with-random-pointer/description/)

用字典记录相关信息,val,next,random分别复制~

```python
class Solution(object):
    def copyRandomList(self, head):
        """
        :type head: RandomListNode
        :rtype: RandomListNode
        """
        if head is None:
            return head
        book={}
        curr=head
        #add val
        while curr:
            book[curr]=RandomListNode(curr.label)
            curr=curr.next
            
        #add next& random
        for i in book:
            if i.next:
                book[i].next=book[i.next]
            if i.random:
                book[i].random=book[i.random]
        return book[head]
```
